## Task 2. Implement [Byte-Pair Encoding(BPE)](https://arxiv.org/pdf/1508.07909.pdf) Tokenizer (80 points)

### Task 2.1. Implementation (60 points)

Implement the tokenizer as the BPETokenizer class:
* Implement `train` method that learns merges and builds the vocabulary of the specified `vocab_size` (25 points).
* Implement `tokenize` method that should tokenize the text according to the learnt merges (25 points).

Your code should have docstrings and comments (10 points).

```python
from typing import List
from collections import defaultdict
import re

class BPETokenizer:
    """
    Byte Pair Encoding (BPE) Tokenizer class for vocabulary construction and text tokenization.

    Parameters:
        - vocab_size (int): The desired vocabulary size.

    Attributes:
        - vocab_size (int): The desired vocabulary size.
        - _vocab (set): The set containing unique tokens in the vocabulary.
        - _eow (str): End-of-word token.
    """

    def __init__(self, vocab_size: int) -> None:
        """
        Args:
            - vocab_size (int): The desired size of the vocabulary.
        """
        self.vocab_size = vocab_size
        self._vocab = set()
        self._eow = '</w>'

    def train(self, corpus: List[str]) -> None:
        """
        Train the tokenizer on a given corpus.

        Parameters:
            - corpus (List[str]): List of strings representing the training data.

        Returns:
            - None
        """
        vocab = self.__get_vocab(corpus)
        self.__add_tokens(vocab)
        # Iteratively merge tokens until vocabulary size is reached
        while len(self._vocab) < self.vocab_size:
            try:
                pair = self.__get_stats(vocab)
            except:
                print('Cannot get stats')
                break
            vocab = self.__merge_vocab(pair, vocab)
            self.__add_tokens(vocab)
        return vocab

    def tokenize(self, text: str) -> List[str]:
        """
        Tokenize the input text using the trained vocabulary.

        Parameters:
            - text (str): The input text to tokenize.

        Returns:
            - List[str]: List of tokens obtained from the input text.
        """
        tokens = []
        i = 0
        while i < len(text):
            found = False
            # Try to find the longest token in the vocabulary
            for j in range(len(text), i, -1):
                if text[i:j] in self._vocab or text[i:j] + self._eow in self._vocab:
                    tokens.append(text[i:j])
                    i = j
                    found = True
                    break
            if not found:
                i += 1
        return tokens
    
    def __add_tokens(self, vocab: dict) -> None:
        """
        Add tokens to the vocabulary.

        Parameters:
            - vocab: The vocabulary to add tokens from.
        """
        for word in vocab:
            tokens = word.split(' ')
            for token in tokens:
                # Ensure vocabulary size limit is not exceeded
                if len(self._vocab) < self.vocab_size:
                    self._vocab.add(token)
    
    def __get_vocab(self, corpus: List[str]) -> dict: 
        """ 
        Given a list of strings, returns a dictionary of words mapping to their frequency  
        count in the data. 

        Parameters:
            - corpus (List[str]): List of strings representing the training data.

        Returns:
            - dict: A dictionary mapping words to their frequency count.
        """
        vocab = defaultdict(int) 
        for line in corpus: 
            # Iterate through each word in the line
            for word in line.split(): 
                # Map the word (converted to a list of characters) to its frequency in the data
                vocab[' '.join(list(word)) + f' {self._eow}'] += 1
        return vocab
        
    def __get_stats(self, vocab) -> tuple[str, str]:
        """
        Get the most frequent pair of symbols in the vocabulary.

        Parameters:
            - vocab: The vocabulary to analyze.

        Returns:
            - Tuple[str, str]: The most frequent pair of symbols.
        """
        pairs = defaultdict(int)
        # Iterate over each word and count the occurrences of symbol pairs
        for word, freq in vocab.items():
            symbols = word.split()
            for j in range(len(symbols) - 1):
                pairs[symbols[j], symbols[j + 1]] += freq
        # Get the pair with the highest frequency
        best = max(pairs, key=pairs.get)
        return best
    
    def __merge_vocab(self, pair, vocab_in) -> dict:
        """
        Merge the most frequent pair of symbols in the vocabulary.

        Parameters:
            - pair (Tuple[str, str]): The pair of symbols to merge.
            - vocab_in: The input vocabulary to merge.

        Returns:
            - dict: The merged vocabulary.
        """
        vocab_out = {}
        # Escape the pair to use it as a regular expression pattern
        bigram = re.escape(" ".join(pair))
        # Create a pattern to find occurrences of the pair
        p = re.compile(r"(?<!\S)" + bigram + r"(?!\S)")
        # Iterate over the input vocabulary and replace occurrences of the pair with the merged result
        for word in vocab_in:
            word_out = p.sub("".join(pair), word)
            vocab_out[word_out] = vocab_in[word]
        return vocab_out
```

### Task 2.2. Analysis on Tweets Dataset (10 points)

Train the BPE tokenizer on the tweets dataset. Try to tokenize the tweets with the tokenizer of different `vocab_size`. For example, train the BPE tokenizer with `vocab_size` of [base_vocab_size, 250, 500, 750, 1000]. Plot the dependency of the average length of the tokenized tweet by `vocab_size` to analyze how `vocab_size` affects the length of the tokenized tweet on average. Tell what `vocab_size` is preferrable and why.

```python
import os


path = '/kaggle/input/nlp-a1-tweets/Assignment1_data'
corpus = []

for name in sorted(os.listdir(path)):
    f = open(path + '/' + name)
    corpus += f.read().split('\n')
    f.close()

len(corpus)
```

```python
base_vocab_size = 125
vocab_sizes = [base_vocab_size, 250, 300, 400, 500, 600, 750, 1000]
bpes = []
average_tokens = []

for vocab_size in vocab_sizes:
    bpes += [BPETokenizer(vocab_size)]
    bpes[-1].train(corpus)
    average_tokens.append(sum([len(bpes[-1].tokenize(s)) for s in corpus]) / len(corpus))

average_tokens
```

```python
import matplotlib.pyplot as plt


plt.plot(vocab_sizes, average_tokens)
plt.xlabel('Vocab Size')
plt.ylabel('Average Number of Tokens')
```

### Answer
> According to the graphical representation, it is evident that beyond a vocabulary size of 500, the rate of decrease in the average number of tokens becomes more gradual. Consequently, based on the elbow rule, a vocabulary size of 500 is considered optimal.
